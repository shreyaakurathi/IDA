# 2.Read the “ratings.csv”file and create a pivot table with index=‘userId’, columns=‘movieId’,values = “rating. 

```python
pivot_movie_user_rating = pd.pivot_table(ratings_df, values="rating", index="userId", columns="movieId")
pivot_movie_user_rating.head()
```

```python
pivot_movie_user_rating.isnull().sum().sum()
```

```python
pivot_movie_user_rating.fillna(value=0, inplace=True)
pivot_movie_user_rating.isnull().sum().sum()
```

# 3.sklearn.metrics.pairwise_distances  can  be  used  to  compute  distancebetween  all  pairs  of users. pairwise_distances() takes a metric parameter for what distance measure touse. Use cosine similarity for finding similarityamong users.Use the following packages.


# 4.from sklearn.metrics import pairwise_distances

# 5.from scipy.spatial.distance import cosine, correlation

```python
users_dist_by_movie = pairwise_distances(pivot_movie_user_rating, metric='cosine')
print(users_dist_by_movie)
```

# 6.Find the 5 most similar user for user with user Id 25.

```python
user_25_similarity = users_dist_by_movie[25 - 1, :]
# to get the top 5 most similar users we can get the highest values' indices
# since the index represents the user ID
five_most_similar_users_to_id_25 = user_25_similarity.argsort()
five_most_similar_users_to_id_25 = five_most_similar_users_to_id_25[:5]
print(f"The users that are most similar to user ID 25 have the ID: {[x+1 for x in five_most_similar_users_to_id_25]} in the same high to low order")
```

# 7.Use the “movies”dataset to find out the names of movies, user 1and user 338 have watched in common and how they have rated each oneof them

```python
rating_movie_df = pd.DataFrame(pd.merge(ratings_df, movies_df, on="movieId"))
rating_movie_df
```

```python
user1_movie = pd.DataFrame(rating_movie_df[rating_movie_df["userId"] == 1][["title","rating"]])
user1_movie.rename(columns = {'title':'title', 'rating':'rating_1'}, inplace = True)
user1_movie.head()
```

```python
user338_movie = pd.DataFrame(rating_movie_df[rating_movie_df["userId"] == 338][["title","rating"]])
user338_movie.rename(columns = {'title':'title', 'rating':'rating_338'}, inplace = True)
user338_movie.head()
```

```python
user2_movie = pd.DataFrame(rating_movie_df[rating_movie_df["userId"] == 2][["title","rating"]])
user2_movie.rename(columns = {'title':'title', 'rating':'rating_2'}, inplace = True)
merged_by_title_user_2_338 = pd.merge(user2_movie, user338_movie, on="title")
merged_by_title_user_2_338.head()
```
